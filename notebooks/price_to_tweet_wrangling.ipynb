```python
import numpy as np
import pandas as pd
```

```python
tsla_df = pd.read_csv('../data/raw/tsla_stock_price.csv')\
                        .drop('Unnamed: 0',axis='columns')
elon_df = pd.read_csv('../data/raw/elonmusk.csv')\
                        .drop('Unnamed: 0',axis='columns')
anomalies_df = pd.read_csv(\
                        '../data/processed/anomaly_tagged_tweet_features.csv')\
                        .drop('Unnamed: 0',axis='columns')
tsla_df['DateTime'] = pd.to_datetime(tsla_df['DateTime'])
elon_df['Time'] = pd.to_datetime(elon_df['Time'])
anomalies_df['Time'] = pd.to_datetime(anomalies_df['Time'])
anomaly_only_df = anomalies_df[anomalies_df['anomalous']!=0.]
no_anomaly_df = anomalies_df[anomalies_df['anomalous']!=1.]
```

```python
tsla_df.columns
```

```python
def nearest(items, pivot):
    return min(items, key=lambda x: abs(x - pivot))
```

```python
print (anomalies_df['Time'].iloc[2000])
nearest(tsla_df['DateTime'],anomalies_df['Time'].iloc[2000])
```

```python
test_df = anomalies_df
```

```python
test_df['stock_time']=test_df['Time'].apply(lambda row: nearest(tsla_df['DateTime'],row))
```

```python
test_df.to_csv('../data/processed/anomalyandstock_tagged_tweet_features.csv')
```

```python
def nearest_price(items, pivot,df):
    timestamp = min(items, key=lambda x: abs(x - pivot))
    return df.loc[df['DateTime']==timestamp,['Open']].values
```

```python
nearest_price(tsla_df['DateTime'],anomalies_df['Time'].iloc[2000],tsla_df)
```

```python
test_df['stock_price']=test_df['Time'].apply(lambda row: nearest_price(tsla_df['DateTime'],row,tsla_df))
```
